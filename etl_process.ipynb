### 3. Limpieza y preparación


Inicialmente reordenamos las columnas del Dataframe con el fin de tener una mejor organización de la información

```python
data_pre = data_pre[[
    "ID",
    "Fecha de la desaparición",
    "Año de la desaparición",
    "Mes de la desaparición",
    "Día de la desaparición",
    "Estado de la desaparición",
    "Clasificación de la desaparición",
    "Entidad que realiza el registro de la desaparición",
    "País donde ocurre la desaparición",
    "Departamento donde ocurre la desaparición DANE",
    "Codigo Dane Departamento",
    "Municipio donde ocurre la desaparición DANE",
    "Codigo Dane Municipio", 
    "Zona donde ocurre la desaparición",
    "Localidad donde ocurre la desaparición",
    "Sexo del desaparecido",
    "País de nacimiento del desaparecido",
    "Estado civil del desaparecido",
    "Grupo mayor y menor de edad del desaparecido",
    "Grupo de edad quinquenal del desaparecido",
    "Grupo de edad judicial del desaparecido",
    "Ciclo vital del desaparecido",
    "Escolaridad del desaparecido",
    "Pertenencia étnica del desaparecido",
    "Pueblo indígena del desaparecido",
    "Pertenencia grupal del desaparecido",
    "Identidad de género del desaparecido",
    "Orientación sexual del desaparecido",
    "Transgénero",
    "Contexto"
]]
```

```python
print(data_pre["ID"].head(10))
print(data_pre["ID"].value_counts())

data_pre.drop(columns = "ID", inplace = True)
# Se elimina la columna
```

```python
print(data_pre["Fecha de la desaparición"].head(10))

def parseDate(x):
    return date(int(x[0:4]), int(x[5:7]), int(x[8:10]))
# Creamos una función para extraer el año, mes y dia

data_pre["Fecha de la desaparición"] = pd.to_datetime(data_pre["Fecha de la desaparición"].apply(lambda x: parseDate(x)))
# Convertimos los datos de la columna a tipo Date (Fecha)
```

```python
print(data_pre["Año de la desaparición"].head(10))
print(data_pre["Año de la desaparición"].value_counts())

data_pre["Año de la desaparición"] = data_pre["Fecha de la desaparición"].apply(lambda x: int(x.year))
# La columna "Año de la desaparición" se llena con el año de la fecha para llenar los datos sin información

data_pre = data_pre[data_pre["Año de la desaparición"] >= 1986]
data_pre = data_pre[data_pre["Año de la desaparición"] <= 2024]
# Se obtienen las desapariciones entre el año 1986 y 2024
```

```python
print(data_pre["Mes de la desaparición"].head(10))

data_pre.drop(columns = "Mes de la desaparición", inplace = True)
# Se elimina la columna
```

```python
print(data_pre["Día de la desaparición"].value_counts())

def getFullWeekday(x):
    if x.weekday() == 0:
        return "Lunes"
    elif x.weekday() == 1:
        return "Martes"
    elif x.weekday() == 2:
        return "Miércoles"
    elif x.weekday() == 3:
        return "Jueves"
    elif x.weekday() == 4:
        return "Viernes"
    elif x.weekday() == 5:
        return "Sábado"
    elif x.weekday() == 6:
        return "Domingo"
# Se crea una función para obtener el dia de la semana a partir de la fecha

data_pre["Día de la desaparición"] = data_pre["Fecha de la desaparición"].apply(lambda x: getFullWeekday(x))
# Se aplica la función a la columna
```

```python
print(data_pre["Estado de la desaparición"].head(10))
print(data_pre["Estado de la desaparición"].value_counts())

data_pre["Estado de la desaparición"].replace({
    "Aparecio Vivo": "Aparecido Vivo",
    "Aparecio Muerto": "Aparecido Muerto"
})
# Se cambian algunos datos
```

```python
print(data_pre["Clasificación de la desaparición"].head(10))
print(data_pre["Clasificación de la desaparición"].value_counts())
```

```python
print(data_pre["Entidad que realiza el registro de la desaparición"].head(10))
print(data_pre["Entidad que realiza el registro de la desaparición"].value_counts())

data_pre.drop(columns = "Entidad que realiza el registro de la desaparición", inplace = True)
# Se elimina la columna
```

```python
print(data_pre["País donde ocurre la desaparición"].head(10))
print(data_pre["País donde ocurre la desaparición"].value_counts())

print("Tamaño antes:", data_pre["País de nacimiento del desaparecido"].shape)

data_pre = data_pre[data_pre["País donde ocurre la desaparición"] == "Colombia"]
# Se filtra para las personas que desaparecieron en Colombia

print("Tamaño después:", data_pre["País de nacimiento del desaparecido"].shape)

data_pre.drop(columns = "País donde ocurre la desaparición", inplace = True)
# Se elimina la columna
```

```python
print(data_pre["Departamento donde ocurre la desaparición DANE"].head(10))
print(data_pre["Departamento donde ocurre la desaparición DANE"].value_counts())
```

```python
print(data_pre["Codigo Dane Departamento"].head(10))
print(data_pre["Codigo Dane Departamento"].value_counts())

data_pre["Codigo Dane Departamento"] = data_pre["Codigo Dane Departamento"].apply(lambda x: int(x))
# Se cambia de tipo float a int el código Dane del Departamento
```

```python
print(data_pre["Municipio donde ocurre la desaparición DANE"].head(10))
print(data_pre["Municipio donde ocurre la desaparición DANE"].value_counts())
```

```python
print(data_pre["Codigo Dane Municipio"].head(10))
print(data_pre["Codigo Dane Municipio"].value_counts())

data_pre.drop(columns = "Codigo Dane Municipio", inplace = True)
# Se elimina la columna
```

```python
print(data_pre["Zona donde ocurre la desaparición"].head(10))
print(data_pre["Zona donde ocurre la desaparición"].value_counts())
```

```python
print(data_pre["Localidad donde ocurre la desaparición"].head(10))
print(data_pre["Localidad donde ocurre la desaparición"].value_counts())

data_pre.drop(columns = "Localidad donde ocurre la desaparición", inplace = True)
# Se elimina la columna
```

```python
print(data_pre["Sexo del desaparecido"].head(10))
print(data_pre["Sexo del desaparecido"].value_counts())
```

```python
print(data_pre["País de nacimiento del desaparecido"].head(10))
print(data_pre["País de nacimiento del desaparecido"].value_counts())

print("Tamaño antes:", data_pre["País de nacimiento del desaparecido"].shape)

data_pre = data_pre[data_pre["País de nacimiento del desaparecido"] == "Colombia"]
# Se filtra para personas nacidas en Colombia

print("Tamaño después:" ,data_pre["País de nacimiento del desaparecido"].shape, "\n")

data_pre.drop(columns = "País de nacimiento del desaparecido", inplace = True)
# Se elimina la columna
```

```python
print(data_pre["Grupo de edad quinquenal del desaparecido"].head(10))
print(data_pre["Grupo de edad quinquenal del desaparecido"].value_counts())

data_pre.drop(columns = "Grupo de edad quinquenal del desaparecido", inplace = True)
# Se elimina la columna
```

```python
print(data_pre["Grupo de edad judicial del desaparecido"].head(10))
print(data_pre["Grupo de edad judicial del desaparecido"].value_counts())

def formatAge(x):
    x1 = x.split("(")
    x2 = x1[1].split(")")
    return f"{x2[0]} Años"
# Se crea una función para obtener el grupo de edad y llevarlo a un formato común

data_pre["Grupo de edad judicial del desaparecido"] = data_pre["Grupo de edad judicial del desaparecido"].apply(lambda x: formatAge(x))
# Se aplica la función a la columna "Grupo de edad judicial del desaparecido"
```

```python
print(data_pre["Ciclo vital del desaparecido"].head(10))
print(data_pre["Ciclo vital del desaparecido"].value_counts())

def getLifeCicle(x):
    range_ = x.split(" ")
    try: 
        inf = int(range_[0])
        sup = int(range_[2])
    except:
        inf = int(range_[0])
        sup = 80

    age = (inf + sup) / 2

    if (age < 6):
        return "Primera infancia (de 0 a 5 años)"
    elif (age >= 6) and (age < 12):
        return "Infancia (de 6 a 11 años)"
    elif (age >= 12) and (age < 18):
        return "Adolescencia (de 12 a 17 años)"
    elif (age >= 18) and (age < 29):
        return "Juventud (de 18 a 28 años)"
    elif (age >= 29) and (age < 60):
        return "Adultez (de 29 a 59 años)"
    elif (age >= 60):
        return "Adulto Mayor (mayor de 60 años)"
# Se crea una función para obtener el ciclo vital del individuo

data_pre["Ciclo vital del desaparecido"] = data_pre["Grupo de edad judicial del desaparecido"].apply(lambda x: getLifeCicle(x))
# Se aplica la función a la columna "Ciclo vital del desaparecido"
```

```python
print(data_pre["Transgénero"].head(10))
print(data_pre["Transgénero"].value_counts())

data_pre.drop(columns = "Transgénero", inplace = True)
# Se elimina la columna
```

```python
print(data_pre["Contexto"].head(10))
print(data_pre["Contexto"].value_counts())

data_pre.drop(columns = "Contexto", inplace = True)
# Se elimina la columna
```

En este proceso de limpieza y preparación se realizaron los siguientes cambios en las columnas de Dataframe:
- ID: Se observa que la columna "ID" contiene números consecutivos del 1 al 195808 y ninguno de estos valores se repite. Sin embargo, estos datos se pueden obtener el índice del Dataframe. Además, al eliminar y/o filtrar algunos datos este "ID" no va a aportar información significativa.
- Fecha de la desaparición: Se observa que esta columna contiene la fecha de la desaparición seguida del carácter "T00:00:00.000". El procedimiento a desarrollar es eliminar este caracter y convertir la fecha a tipo Date.
- Año de la desaparición: En esta columna hacen falta algunos años. Por lo tanto, usamos la columna fecha para obtener este dato. Además, filtramos el Dataframe entre los años 1986 y 2024.
- Mes de la desaparición: Se considera que, aunque la columna aporta el nombre del mes, este puede ser obtenido fácilmente de la columna "Fecha de la desaparición" por lo que se procede a eliminar.
- Dia de la desaparición: Esta columna contiene el dia de la semana que desapareció el individuo. Sin embargo, se observa que algunos datos (129) no contienen información por lo que, usando la columna "Fecha de la desaparición", se extrae el dia de la semana que ocurrió el hecho. Esta información se considera interesante para observar si hay alguna tendencia respecto al dia de la semana que desaparece una persona.
- Estado de la desaparición: En esta columna aparece si el individuo Apareció vivo, muerto o sigue desaparecido. Aqui solamente se reemplazan los valores "Aparecio Vivo" y "Aparecio Muerto" por "Aparecido Vivo" y "Aparecido Muerto", respectivamente, para mejor claridad de la información.
- Clasificación de la desaparición: Esta columna se conserva ya que se realizará un análisis general de desaparecidos y un análisis de desapariciones presuntamente forzosas.
- Entidad que realiza el registro de la desaparición: Esta columna no se considera relevante para el análisis por lo que se procede a eliminar.
- País donde ocurre la desaparición: En esta columna se conservan los registros de los individuos desaparecidos en Colombia. Al filtrar para Colombia el Dataset disminuye de 189934 a 188639 registros. Luego, se elimina la columna.
- Departamento donde ocurre la desaparición DANE: Esta columna se conserva ya que aporta información de interés de los departamentos donde se han reportado desapariciones.
- Codigo Dane Departamento: Los datos de esta columna se utilizarán para realizar un mapa donde se identificarán el número de desapariciones por departamento por periodo presidencial. Se cambia a un dato de tipo entero.
- Municipio donde ocurre la desaparición DANE: Esta columna se conserva para hacer un análisis de los municipios con más desapariciones en colombia.
- Codigo Dane Municipio: Esta columna se elimina ya que para el análisis basta con saber el nombre del municipio.
- Zona donde ocurre la desaparición: Esta columna se conserva para el análisis.
- Localidad donde ocurre la desaparición: Se elimina esta columna ya que no se realizará un análisis de desapariciones en Bogotá.
- Sexo del desaparecido: Se conserva para el análisis.
- País de nacimiento del desaparecido: Se filtra el Dataset por "colombia" y se elimina esta columna. Al filtrar el Dataset Dsiminuye de 188639 a 172952 registros.
- Estado civil del desaparecido: Esta columna se considera útil para el análisis, por lo tanto, se conserva.
- Grupo mayor y menor de edad del desaparecido: Se elimina porque no se conserva relevante para el análisis.
- Grupo de edad quinquenal del desaparecido: Se considera que tiene información muy parecida a las columnas "Grupo mayor y menor de edad del desaparecido" y "Grupo de edad judicial del desaparecido". Por ende, se elimina esta columna.
- Grupo de edad judicial del desaparecido: Se conserva esta columna llevando todas las edades a un formato común.
- Ciclo vital del desaparecido: Se conserva ya que puede realizarse un análisis al rededor de esto.
- Escolaridad del desaparecido: De esta columna se puede extraer un análisis, por lo tanto, se conserva.
- Pertenencia étnica del desaparecido: Esta columna se conserva ya que se considera de interés conocer esta variable en las desapariciones.
- Pueblo indígena del desaparecido: Se elimina esta columna ya que hay muy pocos datos de interés.
- Pertenencia grupal del desaparecido: Esta columna se elimina ya que no se realizará un análisis relacionado a este tema.
- Identidad de género del desaparecido: Se agrupan los "Transexuales" y "No Aplica" como LGBTIQ+.
- Orientación sexual del desaparecido: Se elimina esta columna ya que no hay suficientes datos relevantes para realizar este análisis.
- Transgénero: Se elimina ya que no hay suficiente información de la cual extraer un análisis.
- Contexto: Esta columna contiene un solo dato durante todo el Dataset. Al no ofrecer nueva información se elimina esta columna.

```python
def calcularPeriodoPres(x):
    if (x >= 1986) and (x < 1990):
        return "1986-1990 | Virgilio Barco"
    elif (x >= 1990) and (x < 1994):
        return "1990-1994 | César Gaviria"
    elif (x >= 1994) and (x < 1998):
        return "1994-1998 | Ernesto Samper"
    elif (x >= 1998) and (x < 2002):
        return "1998-2002 | Nicolás Pastrana"
    elif (x >= 2002) and (x < 2006):
        return "2002-2006 | Álvaro Uribe 1"
    elif (x >= 2006) and (x < 2010):
        return "2006-2010 | Álvaro Uribe 2"
    elif (x >= 2010) and (x < 2014):
        return "2010-2014 | Juan Manuel Santos 1"
    elif (x >= 2014) and (x < 2018):
        return "2014-2018 | Juan Manuel Santos 2"
    elif (x >= 2018) and (x < 2022):
        return "2018-2022 | Iván Duque"
    elif (x >= 2022) and (x <= 2024):
        return "2022-2024 | Gustavo Petro"
# Se crea una función para calcular el periodo presidencial

data_pre["Periodo presidencial"] = data_pre["Año de la desaparición"].apply(lambda x: calcularPeriodoPres(x))
# Se aplica la función a la columna "Periodo presidencial"
data_pre["Periodo presidencial"].value_counts()
```

Se creó una columna llamada "Periodo presidencial" para realizar el análisis en base a estos periodos.
